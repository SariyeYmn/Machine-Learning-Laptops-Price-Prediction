```python
# İşlemci sütununu ikiye ayıran fonksiyon
def split_processor(row):
    processor_info = row["islemci"]
    
    # Düzenli ifade ile işlemci ve sürümü ayır
    intel_match = re.match(r"Intel\s+(Celeron|Core i[0-9]+)", processor_info)
    amd_match = re.match(r"AMD\s+(Ryzen [0-9]+)", processor_info)
    apple_match = re.match(r"Apple\s+(M[1-2](?:\sPro)?)", processor_info)
    
    if intel_match:
        processor = "Intel " 
        processor_version = intel_match.group(1)
    elif amd_match:
        processor = "AMD " 
        processor_version = amd_match.group(1)
    elif apple_match:
        processor = "Apple " 
        processor_version = apple_match.group(1)
    else:
        processor = ""
        processor_version = ""
    
    return pd.Series([processor, processor_version])
# İşlemci sütununu ikiye bölen yeni sütunları ekleyin
data[["islemci_markasi", "islemci_modeli"]] = data.apply(split_processor, axis=1)
```

```python
# Encoding yapacak fonksiyon
def encode_ekran_karti(ekran_karti):
    if 'Intel' in ekran_karti:
        return 'Intel'
    elif 'AMD' in ekran_karti:
        return 'AMD'
    elif 'Apple' in ekran_karti:
        return 'Apple'
    elif 'NVIDIA' in ekran_karti:
        return 'NVIDIA'
    elif 'Onboard' in ekran_karti:
        return 'Onboard'
    else:
        return 'Diğer' 

# Encoding fonksiyonunu uygula
data['ekran_karti_encoded'] = data['ekran_karti'].apply(encode_ekran_karti)

print(data[['ekran_karti','ekran_karti_encoded']])
```
