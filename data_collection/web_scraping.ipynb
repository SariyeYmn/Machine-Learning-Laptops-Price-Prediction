```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

headers = {
    "User-Agent": "Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Mobile Safari/537.36"
}

site_url = "https://www.teknosa.com"
ssdler = []
ekran_boyutlari = []
cozunurlukler = []
model_kodlari = []
hddler = []
islemciler = []
ekran_kartlari = []
ramler = []
urun_adlari = []
fiyatlar=[]
markalar=[]

for i in range(20):
    anasayfa_url = "https://www.teknosa.com/laptop-notebook-c-116004?page=" + str(i)
    response = requests.get(anasayfa_url, headers=headers)
    soup = BeautifulSoup(response.content, "html.parser")
    urunler = soup.find_all("div", attrs={"id":"product-item"})
    
    #print(urunler)

    for urun in urunler:
        urun_fiyat=urun.find("div",class_="prd-prc2").text.strip()
        #print(urun_fiyat)
        
        urun_adi = urun.find("h3", class_="prd-title prd-title-style").text.strip()
        #print(urun_adi)
        urun_linki = urun.find("a")
    
        print(urun_linki)
        if urun_linki:
            link_adresleri = site_url + urun_linki.get('href')
            print(link_adresleri)
            responser_2 = requests.get(link_adresleri, headers=headers)
            soup_2 = BeautifulSoup(responser_2.content, "html.parser")
            urun_ozellikleri = soup_2.find_all("div", class_="ptf-item")
            
            p=soup_2.find("div",class_="pdp-base")
            
            if p :
                urun_marka=p.find("a",class_="link").text
                print(urun_marka)
            else:
                print("bulunmadı")
            urun_ozellik_dict = {}
            for ozellik in urun_ozellikleri:
                tablo = ozellik.find_all("th")
                tablo_degerler = ozellik.find_all("td", class_=False)

                for i in range(min(len(tablo), len(tablo_degerler))):
                    tablo_text = tablo[i].text.strip()
                    deger_text = tablo_degerler[i].text.strip()
                    urun_ozellik_dict[tablo_text] = deger_text

                    # Liste değerlerini ekle
                    if tablo_text == "SSD Kapasitesi":
                        ssdler.append(deger_text)
                    elif tablo_text == "Ekran Boyutu":
                        ekran_boyutlari.append(deger_text)
                    elif tablo_text == "Ekran Çözünürlüğü (Piksel)":
                        cozunurlukler.append(deger_text)
                    elif tablo_text == "Model Kodu":
                        model_kodlari.append(deger_text)
                    elif tablo_text == "HDD Kapasitesi":
                        hddler.append(deger_text)
                    elif tablo_text == "İşlemci":
                        islemciler.append(deger_text)
                    elif tablo_text == "Ekran Kartı Modeli":
                        ekran_kartlari.append(deger_text)
                    elif tablo_text == "Ram":
                        ramler.append(deger_text)
                        
            urun_adlari.append(urun_adi)
            fiyatlar.append(urun_fiyat)
            markalar.append(urun_marka)

            # Eğer içteki for döngüsünden çıkıldıktan sonra listenin uzunluğu değişmediyse, eksik olan listelere -1 ekle
            if len(ssdler) != len(urun_adlari):
                ssdler.append("0")
            if len(ekran_boyutlari) != len(urun_adlari):
                ekran_boyutlari.append("0")
            if len(cozunurlukler) != len(urun_adlari):
                cozunurlukler.append("0")
            if len(model_kodlari) != len(urun_adlari):
                model_kodlari.append("0")
            if len(hddler) != len(urun_adlari):
                hddler.append("0")
            if len(islemciler) != len(urun_adlari):
                islemciler.append("0")
            if len(ekran_kartlari) != len(urun_adlari):
                ekran_kartlari.append("0")
            if len(ramler) != len(urun_adlari):
                ramler.append("0")



df = pd.DataFrame({
    "ÜRÜN ADI": urun_adlari,
    "MODEL KODU": model_kodlari,
    "SSD": ssdler,
    "EKRAN BOYUTU": ekran_boyutlari,
    "ÇÖZÜNÜRLÜK": cozunurlukler,
    "HDD": hddler,
    "İŞLEMCİ": islemciler,
    "EKRAN KARTI": ekran_kartlari,
    "RAM": ramler,
    "FİYAT": fiyatlar,
    "MARKA": markalar
})

#df.to_csv("teknosa_veri1.csv" , index=False)
df.to_excel("teknosa_veri1.xlsx", index=False)
```
